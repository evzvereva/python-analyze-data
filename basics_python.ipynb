```python
"""
Задание 1
Даны 2 переменные, в которых хранятся строки произвольной длины: phrase_1 и phrase_2.
Напишите код, который проверяет какая из этих строк длиннее.
"""


def check_length_strings():
    """
    С помощью функции осуществляет запрос фраз. Возвращает результат проверки какая из этих строк длиннее.
    """
    phrase_1 = input('Введите фразу 1: ')
    phrase_2 = input('Введите фразу 2: ')

    if len(phrase_1) > len(phrase_2):
        print('Фраза 1 длиннее фразы 2')
    elif len(phrase_1) < len(phrase_2):
        print('Фраза 2 длиннее фразы 1')
    elif len(phrase_1) == len(phrase_2):
        print('Фразы равной длины')
    else:
        print('Проверьте на корректность введенных данных')


check_length_strings()
```

```python
""" 
Задание 2
Дана переменная, в которой хранится число (год). Необходимо написать программу, которая выведет, является ли данный год високосным или обычным.
"""

def get_data_determine_leap_year():
    """
    С помощью функции осуществляется запрос данных о годе, возвращает результат високсного/обычного года
    """
    year = int(input('Введите год: '))

    if year % 4 == 0 or year % 400 == 0:
        print('Високосный год')
    elif year % 4 != 0 or year % 400 != 0:
        print('Обычный год')
    else:
        print('Проверьте на корректность введенных данных')


get_data_determine_leap_year()
```

```python
"""
Задание 4
Вам нужно написать программу для подбора упаковок по размерам товара. Размеры (ширина, длина, высота) хранятся в переменных (в сантиметрах):

Используйте следующие правила:

если каждое из трех измерений менее или равно 15 сантиметрам, то выведите на экран "Коробка №1";
если хотя бы одно из измерений больше 15 сантиметров, но менее 50 сантиметров, то выводите "Коробка №2";
если длина товара больше 2 метров, то выводите "Упаковка для лыж";
во всех остальных случаях выводите "Стандартная коробка №3".
"""


def request_product_parameters():
    """
    С помощью функции запрашиваются данные о размерах товара у пользователя. В результате полученных данных 
    осуществляется вызов функции get_selection_pack(width, length, height)
    """
  
    width = int(input('Введите ширину товара в сантиметрах: '))
    length = int(input('Введите длину товара в сантиметрах: '))
    height = int(input('Введите высоту товара в сантиметрах: '))

    def get_selection_pack(width, length, height):
        """
        Функция принимает параметры (ширина, длина, высота), возвращает результат подбора упаковки для товара
        """
        if width <= 15 and length <= 15 and height <= 15:
            print('Коробка №2')
        elif 15 < width < 50 or \
                15 < length < 50 or \
                15 < height < 50:
            print('Коробка №2')
        elif length > 200:
            print('Упаковка для лыж')
        else:
            print('Стандартная коробка №3')

    return get_selection_pack(width, length, height)


request_product_parameters()
```

```python
"""
Задание 5 (необязательное)
Дана переменная, в которой хранится шестизначное число (номер проездного билета). Напишите программу,
которая будет определять, является ли данный билет "счастливым". Билет считается счастливым,
если сумма первых трех цифр совпадает с суммой последних трех цифр номера.

Примеры работы программы:

number = 123456
Результат:
Несчастливый билет

number = 123321
Результат:
Счастливый билет
"""


def get_number():
    """
    Функция позволяет запросить шестизначное число у пользователя и вернуть результат (счастливый/несчастливый
    билет). Суммируются отдельно первые три значения и отдельно последние три значения целого шесстизначного числа, 
    сравнив в дальнейшем результат для проверки равны между собой или нет.
    """
    number = int(input('Введите целое шестизначное число: '))

    list_number = [int(n) for n in str(number)]  # список чисел из целого числа, полученные из переменной "number"
    sum_one_num = sum(list_number[:3])  # суммирование первых трех чисел целого числа
    sum_two_num = sum(list_number[3:])  # суммирование последних трех чисел целого числа

    # Проверка по условию

    if sum_one_num == sum_two_num:
        print('Счастливый билет')
    elif sum_one_num != sum_two_num:
        print('Несчастливый билет')
    else:
        print('Проверьте на корректность введенных данных')


get_number()
```

```python
"""
Задание 6 (необязательное)
Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник). Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

если пользователь выбрал круг, запрашиваем его радиус,
если треугольник – длины трех его сторон;
если прямоугольник – длины двух его сторон.
Пример работы программы:

Введите тип фигуры:
Круг

Введите радиус круга:
10
Результат:
Площадь круга: 314.16

Введите тип фигуры:
Треугольник

Введите длину стороны A:
2

Введите длину стороны B:
2

Введите длину стороны C:
3
Результат:
Площадь треугольника: 1.98
"""

import math  # импортируем модуль, который позволяет работать с числами


def get_data_figure():
    """
    Функция позволяет получить от пользователя название фигуры и возвращает результат расчета ее
    площади с помощью вложенной функции "get_calculation_area_figure(type_figure)".
    """
    type_figure = input('Введите тип фигуры с заглавной буквы (Круг/Треугольник/Прямоугольник): ')

    def get_calculation_area_figure(type_figure):
        """
        Функция принимает параметр тип фигуры, возвращает результат расчета ее площади на основании запрощенных
        дополнительных данных (длину сторон/радиус и т.д. в зависимости от типа фигуры) от пользователя.
        """
        if type_figure == 'Круг':
            radius = int(input('Введите радиус круга: '))
            area_circle = round(((radius ** 2) * 3.14), 2)
            print(f'Площадь круга:  {area_circle} кв. см')
        elif type_figure == 'Треугольник':
            len_a = int(input('Введите длину стороны A: '))
            len_b = int(input('Введите длину стороны B: '))
            len_c = int(input('Введите длину стороны C: '))
            half_meter = (len_a + len_b + len_c) / 2  # получаем полупериметр треугольника
            area_triangle = round(math.sqrt(half_meter * (half_meter - len_a) * (half_meter - len_b) * (half_meter - len_c)), 2)  # получаем площадь треугольника, результат 6округлен до десятых
            print(f'Площадь треугольника: {area_triangle} кв. см')
        elif type_figure == 'Прямоугольник':
            len_a = int(input('Введите длину стороны A: '))
            len_b = int(input('Введите длину стороны В: '))
            area_rectangle = round((len_a * len_b), 2)  # получаем площадь прямоугольника
            print(f'Площадь прямоугольника: {area_rectangle} кв. см')
        else:
            print('Проверьте на корректность введенных данных')

    return get_calculation_area_figure(type_figure)


get_data_figure()
```
